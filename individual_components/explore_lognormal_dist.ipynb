```python
import numpy as np
import matplotlib.pyplot as plt

mean_space = np.linspace(-1, 0, 10)
std = 0.13241694217637887

fig, ax = plt.subplots(1, 1)
for mean in mean_space:
    x = np.linspace(0.0001, 1, 1000)
    y = 1/(x * std * np.sqrt(2 * np.pi)) * np.exp(-0.5 * ((np.log(x) - mean) / std)**2)
    ax.plot(x, y, label=f'mean={mean:.2f}')
    
ax.legend()

    
```
