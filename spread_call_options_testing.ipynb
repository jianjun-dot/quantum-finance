```python
import matplotlib.pyplot as plt

data = pd.read_csv('spread_call_test_2024-01-05 18:10:51.csv', names=["exact", "estimated", "lower bound", "upper bound"])
y_error = []
lower_bounds = data["lower bound"]
upper_bounds = data["upper bound"]
for index in range(len(lower_bounds)):
    y_error.append([data["estimated"][index] - lower_bounds[index], upper_bounds[index]-data["estimated"][index]])

print(y_error)

mean = np.mean(data["estimated"])

plt.plot(range(len(data['exact'])), data["exact"], 'r')
plt.axhline(y=mean, color='b', linestyle='-')
plt.errorbar(range(len(data['estimated'])), data["estimated"], yerr=np.array(y_error).T, fmt='o')
```
